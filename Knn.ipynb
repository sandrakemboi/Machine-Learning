```python
 #EDA
import pandas as pd
import numpy  as np#Uses a number array
import sklearn
#Used to train algorithim and standardize data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

#Testing the model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
data_df = pd.read_csv('diabetes.csv')
data_df.info()
data_df.shape
data_df.head()
```

```python
 #Dealing with col with zero values.Some ML models such as regression and KNN do not perform well with missing values.
zero_not = data_df[["Glucose","BloodPressure","SkinThickness","BMI","Insulin"]]

for column in zero_not:
    data_df[column]= data_df[column].replace(0, np.nan)
    mean = int(data_df[column].mean(skipna=True))
    data_df[column]=data_df[column].replace(np.nan, mean)
```

Splitting the dataset

```python
x = data_df.iloc[:,0:8]#X the features used to predict something
y = data_df.iloc[:,8]#y the target/label feature one is predicting
x_train, x_test,y_train,y_test = train_test_split(x,y,random_state=0,test_size=0.2)
```

Feature scaling
NB:When using any algorithim that computes distance or assumes normality ,always scale your features eg KNN,SVM,Logistic regression

```python
sc_x= StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

Define the model and build the model

```python
import math
math.sqrt(len(y_train))#KNN does not work with even numbers
```

```python
#Define the model:init KNN
classifier = KNeighborsClassifier(n_neighbors=11, p=2,metric='euclidean')
```

```python
#train model
classifier.fit(x_train,y_train)
```

MODEL EVALUATION

```python
#Accuracy and f1
y_pred = classifier.predict(x_test)
y_pred
```

```python
#Evaluate the model
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(f1_score(y_test,y_pred))
```

Generally data scientists always ask for f1 scores, but the gebneral public always asks for accuracy.
F1 score is always more accurate.

```python
#Accuracy score
accuracy_score(y_test,y_pred)
```
